```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

data =pd.read_csv('IPL Matches 2008-2020.csv')
data.head()
```

```python
data.shape #counting row and column of dataset
```

```python
data['player_of_match'].value_counts()[0:5] #counting the value of player of the match top 5(0to4)
```

```python
plt.figure(figsize=(8,5)) #plt.bar(categorical values(x) ,numerical values(y))
plt.bar(list(data['player_of_match'].value_counts()[0:5].keys()),list(data['player_of_match'].value_counts()[0:5]),color='skyblue')

plt.title('ipl stats')
plt.xlabel('player of the match')
plt.ylabel('Runs')
plt.show()
```

```python
list(data['result'].value_counts().keys())
```

```python
list(data['result'].value_counts().keys())
```

```python
data['toss_winner'].value_counts()[0:5]
```

```python
batting_first=data[data['result']!=0] #condition where number assigning a variable
```

```python
batting_first.head()
```

```python
#making histogram use for visualization of the numerical data
plt.figure(figsize=(7,5))
plt.hist(batting_first['result_margin'])
plt.title('distribution of runs')
plt.show()
```

```python
#finding out the number of wins 
batting_first['winner'].value_counts()
```

```python
#making bar plot for batting first team

plt.figure(figsize=(12,5))
plt.bar(list(batting_first['winner'].value_counts()[0:5].keys()),list(batting_first['winner'].value_counts()[0:5]),color=['blue','yellow','purple','red','pink'])
plt.title('IPL')
plt.xlabel('Teams')
plt.ylabel('no of wins')
plt.show()
```

```python
#making pie(parameter(numerical),parameter(categorical))
plt.figure(figsize=(8,8))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
#extracting those records where a team has won after batting second
batting_second=data[data['result']!='runs']    #making of variable dataframe

batting_second.head()
```

```python


batting_second.value_counts()

#making barplot of team batting_second
plt.figure(figsize=(11,5))
plt.bar(list(batting_second['winner'].value_counts()[0:5].keys()),list(batting_second['winner'].value_counts()[0:5]),color=['purple','blue','red','pink','yellow'])
plt.title('IPL')
plt.xlabel('Teams win by wickets')
plt.ylabel('no of matches')
plt.show()
```
